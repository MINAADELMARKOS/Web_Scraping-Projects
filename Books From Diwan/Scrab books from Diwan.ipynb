```python
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd
from itertools import zip_longest
```

```python
start=1
src = requests.get(f"https://diwanegypt.com/product-category/books/english-books/business/page/{start}/")
soup=BeautifulSoup(src.text,"html.parser")
```

```python
# books=soup.find_all("ul",{"class":"products columns-5"})
```

```python
titles=[]
authors=[]
prices=[]
images=[]

data=[]

for index,book in enumerate(soup.find("ul",{"class":"products columns-5"}).find_all("li")):

  title=book.find("h2",{"class":"woocommerce-loop-product__title"}).text
  author=book.find("span",{"class":"author"}).text
  price=book.find("span",{"class":"woocommerce-Price-amount amount"}).text
  imag=book.find("a",{"class":"woocommerce-LoopProduct-link woocommerce-loop-product__link"}).find("img").get("src")

  titles.append(title)
  authors.append(author)
  prices.append(price)
  images.append(imag)
  data.append([title,author,price,imag])

diwan_data = pd.DataFrame(data, columns=["title","author","price","imag"])

diwan_data1 = pd.DataFrame(
    {'title': titles,
     'author': authors,
     'price': prices,
     'imag':images
    })
```

```python
with open("vd.csv","w") as fid:
  wr=csv.writer(fid)
  wr.writerow(["title","author","price","imag"])
  wr.writerows(data)
```
